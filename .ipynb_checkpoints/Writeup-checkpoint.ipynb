# PID Controller

## Introduction

The purpose of this project is make a PID Controller for Udacity simulator. The P, I and D factors were also tuned so the car in the simulator without driving off the road.  The goal is to try to reduce the cte (Cross Track Error) as much as possible so that the car stays in the middle of the road.

To evaluate different, the PID values, the number of steps and the name of an output file were added as arguments for the pid executable.  The output files were csv files with the step and cte values.  These csv were used to create graphs to help visualize the effect of different PID values.

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def merge_data(data_pd_arr):
    merged_pd = pd.merge(data_pd_arr[0], data_pd_arr[1], on='step', how='inner')
    
    for i in range(2, len(data_pd_arr)):
        merged_pd = pd.merge(merged_pd, data_pd_arr[i], on='step', how='inner')
        
    return merged_pd
```

```python
def draw_graph(data_pd):
    x = data_pd['step']
    y_colnames = [colname for colname in data_pd.columns if colname != 'step']

    colors = ['red', 'green', 'orange', 'blue', 'yellow', 
              'purple', 'aqua', 'fuchsia', 'goldenrod']
    
    plt.axhline(color='black')

    for i in range(len(y_colnames)):
        plt.plot(x, data_pd[y_colnames[i]], color=colors[i])
    
    plt.legend()
    plt.ylabel('cte')
    plt.xlabel('step')

    plt.show()    
```

## Tuning

### P - proportional factor

With only the P factor set, the cte should oscillate around 0.  P should be negative because we want to move proportionally in the opposite direction of the cte to try to correct the error.  We'll compare the following P factors for 500 steps.

```python
out_500_075_pd = pd.read_csv('cte_data/out_500_075.csv', names=['step', 'P=-0.75'])
out_500_100_pd = pd.read_csv('cte_data/out_500_100.csv', names=['step', 'P=-1.00'])
out_500_125_pd = pd.read_csv('cte_data/out_500_125.csv', names=['step', 'P=-1.25'])
```

```python
out_500_pd = merge_data([out_500_075_pd, out_500_100_pd, out_500_125_pd])
```

```python
draw_graph(out_500_pd)
```

The best P value out of the 3 is -1.00.  The controller oscillates closer to 0 for this P value.

### D - differential factor

We will use the best P value of -1.00 and compare D values from -50.0 to 0.0 for 2000 steps.

```python
d_vals = ['0000', '1000', '2000', '3000', '4000', '5000']

out_2000_100_pd_arr = []
for d_val in d_vals:
    csv_file = 'cte_data/out_2000_100_%s.csv' % d_val
    d_val_f = float(d_val)/-100.0
    col_name = 'P=-1.00, D=%.2f' % d_val_f

    data_pd = pd.read_csv(csv_file, names=['step', col_name])
    out_2000_100_pd_arr.append(data_pd)
```

```python
out_2000_100_pd  = merge_data(out_2000_100_pd_arr)
```

```python
draw_graph(out_2000_100_pd)
```

For D=0.0, the controller diverges from the y-axis at around 800 steps.  The other D values seem to converge closely to 0. 

Here we will remove D=0.0 and add D=-60.00 and D=-70.00 and increase the number of steps to 3000.

```python
d_vals = ['1000', '2000', '3000', '4000', '5000', '6000', '7000']

out_3000_100_pd_arr = []
for d_val in d_vals:
    csv_file = 'cte_data/out_3000_100_%s.csv' % d_val
    d_val_f = float(d_val)/100.0
    col_name = 'P=-1.00, D=-%.2f' % d_val_f    

    data_pd = pd.read_csv(csv_file, names=['step', col_name])
    out_3000_100_pd_arr.append(data_pd)
```

```python
out_3000_100_pd  = merge_data(out_3000_100_pd_arr)
```

```python
draw_graph(out_3000_100_pd)
```

Clearly, we can eliminate D=-10.00 and D=-30.00 because of their large spikes.  Let's repeat this with 4000 steps.

```python
d_vals = ['2000', '4000', '5000', '6000', '7000']

out_4000_100_pd_arr = []
for d_val in d_vals:
    csv_file = 'cte_data/out_4000_100_%s.csv' % d_val
    d_val_f = float(d_val)/-100.0
    col_name = 'P=-1.00, D=%.2f' % d_val_f   

    data_pd = pd.read_csv(csv_file, names=['step', col_name])
    out_4000_100_pd_arr.append(data_pd)
```

```python
out_4000_100_pd  = merge_data(out_4000_100_pd_arr)
```

```python
draw_graph(out_4000_100_pd)
```

We can remote D=-20.0 and D=-70.0.

```python
d_vals = ['4000', '5000', '6000']

out_4000_100_pd_arr = []
for d_val in d_vals:
    csv_file = 'cte_data/out_4000_100_%s.csv' % d_val
    d_val_f = float(d_val)/-100.0
    col_name = 'P=-1.00, D=%.2f' % d_val_f   

    data_pd = pd.read_csv(csv_file, names=['step', col_name])
    out_4000_100_pd_arr.append(data_pd)
```

D=-50.0 appears to be the best out of the 3 factors.

### I - integral factor

To try to fine tune the steering of the car, we will look at I values from -0.04 to 0.0 for 5000 steps.

```python
d_vals = ['000', '001', '002', '003', '004']

out_5000_100_5000_pd_arr = []
for d_val in d_vals:
    csv_file = 'cte_data/out_5000_100_5000_%s.csv' % d_val
    d_val_f = float(d_val)/-100.0
    col_name = 'P=-1.00, D=-50.00, I=%.2f' % d_val_f   

    data_pd = pd.read_csv(csv_file, names=['step', col_name])
    out_5000_100_5000_pd_arr.append(data_pd)
```

```python
out_5000_100_5000_pd  = merge_data(out_5000_100_5000_pd_arr)
```

```python
draw_graph(out_5000_100_5000_pd)
```

Clearly, we can eliminate I=0.0 and I=-0.01.  Let's look at I from -0.04 to -0.02 at 7000 steps.

```python
d_vals = ['002', '003', '004']

out_7000_100_5000_pd_arr = []
for d_val in d_vals:
    csv_file = 'cte_data/out_7000_100_5000_%s.csv' % d_val
    d_val_f = float(d_val)/-100.0
    col_name = 'P=-1.00, D=50.00, I=%.2f' % d_val_f   

    data_pd = pd.read_csv(csv_file, names=['step', col_name])
    out_7000_100_5000_pd_arr.append(data_pd)
```

```python
out_7000_100_5000_pd  = merge_data(out_7000_100_5000_pd_arr)
```

```python
draw_graph(out_7000_100_5000_pd)
```

The best I value out of the 3 is -0.03.

## PID

Through trial and error, we found a reasonably good PID controller with the values of P=-1.00, D=-50.00, I=-0.03.

```python
y_colname = 'P=-1.00, D=-50.00, I=-0.03'
data_pd = pd.read_csv('cte_data/out_10000_100_5000_003.csv', 
                      names=['step', y_colname])

x = data_pd['step']

plt.axhline(color='black')
plt.plot(x, data_pd[y_colname], color='blue')
    
plt.legend()
plt.ylabel('cte')
plt.xlabel('step')

plt.show()    
```
